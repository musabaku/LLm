## 2.Yapay Sinir Ağı Modelinin Eğitilmesi

Tensorflow'da tek hidden layer'lı bir yapay sinir ağı modeli oluşturuldu. Data oldukça az olduğu için bir deep learning modeli başarılı olamazdı. Ev sahibi takım başarı vektörü ve maç sonuçlarında (11,1) lik bir matris ile deplasman takımının (1,11) lik matrisini çaprtığımızda (1,121)'lik bir input matrisi elde ediyoruz. Bu matris iki takımın dünya kupasındaki başarıları ve grup maçlarına göre bir ilişki bilgisi içeriyor. 

Maçlarda yaşanabilecek tahmin edilemeyecek durumlar için grup maç sonuçlarına beyaz gürültüde ekleniyor.

```python
import tensorflow as tf
import numpy as np
import matplotlib.pylab as plt

results = []

# Model parameters
W1 = tf.Variable(tf.random_normal([121, 80], stddev=0.07), name='W1')
b1 = tf.Variable(tf.random_normal([80]), name='b1')
# and the weights connecting the hidden layer to the output layer
W2 = tf.Variable(tf.random_normal([80, 20], stddev=0.07), name='W2')
b2 = tf.Variable(tf.random_normal([20]), name='b2')
# and the weights connecting the hidden layer to the output layer
W3 = tf.Variable(tf.random_normal([80, 2], stddev=0.07), name='W3')
b3 = tf.Variable(tf.random_normal([2]), name='b3')


# Model input and output
xd = tf.placeholder(tf.float32, [1, 121])

hidden_out = tf.add(tf.matmul(xd, W1), b1)
hidden_out = tf.nn.relu(hidden_out)

# Dropout on hidden layer: RELU layer
keep_prob = tf.placeholder(tf.float32)
hidden_out_dropout = tf.nn.dropout(hidden_out, keep_prob)

hidden_out2 = tf.add(tf.matmul(hidden_out_dropout, W2), b2)
hidden_out2 = tf.nn.relu(hidden_out2)

# Dropout on hidden layer: RELU layer
keep_prob2 = tf.placeholder(tf.float32)
hidden_out2_dropout = tf.nn.dropout(hidden_out2, keep_prob2)

#linear_model = W * x + b
y = 1*(tf.matmul(hidden_out_dropout, W3) + b3)
    #y = tf.nn.relu(1*(tf.matmul(hidden_out2, W3) + b3) + 1*(hidden_out_wide))
    #y = 1 * tf.matmul((tf.matmul(hidden_out2, W3) + b3), Wy1) + 1 * tf.matmul((hidden_out_wide), Wy2)

y_ = tf.placeholder(tf.float32, [1, 2])

# loss
loss = tf.reduce_sum(tf.square(y - y_)) / len(world_data)

# optimizer
optimizer = tf.train.GradientDescentOptimizer(0.007).minimize(loss)

#Start session
sess = tf.InteractiveSession(config=tf.ConfigProto(log_device_placement=True))
tf.global_variables_initializer().run()

# training loop
for i in range(40):
    cost2 = 0
    # training data
    for n in range(len(world_data)):
        mu, sigma = 0, 0.11  # mean and standard deviation
        
        x_train_home_success = np.array(world_data[n]['homeS']).reshape(5, 1)
        x_train_home_group = np.array(world_data[n]['goalH']).reshape(6, 1)
        s = np.random.normal(mu, sigma)
        x_train_home_group = s * x_train_home_group + x_train_home_group

        x_train_away_success = np.array(world_data[n]['awayS']).reshape(1, 5)
        x_train_away_group = np.array(world_data[n]['goalA']).reshape(1, 6)
        s = np.random.normal(mu, sigma)      
        x_train_away_group = s * x_train_away_group + x_train_away_group


        x_train_home = np.append(x_train_home_success, x_train_home_group).reshape(11, 1)
        x_train_away = np.append(x_train_away_success, x_train_away_group).reshape(1, 11)

        x_train_d = np.matmul(x_train_home, x_train_away)
        x_train_d = np.reshape(x_train_d, (1, 121))

        y_train_home = np.array(world_data[n]['homeG']).reshape(1, 1)
        y_train_away = np.array(world_data[n]['awayG']).reshape(1, 1)
        y_train = np.append(y_train_home, y_train_away).reshape(1, 2)


        _, cost = sess.run([optimizer, loss], feed_dict={xd: x_train_d, y_: y_train, keep_prob :0.75, keep_prob2 : 0.85})
        cost2 = (i*cost2 + cost) / (i + 1)

        results.append(cost2)

# Add ops to save and restore all the variables.
saver = tf.train.Saver(save_relative_paths=True)
#saver = tf.train.Saver()

# Save the variables to disk.
save_path = saver.save(sess, "worldcup(copy)/model.ckpt")
```
